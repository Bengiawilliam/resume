```python
import ta
```

```python
data['rsi'] = ta.momentum.RSIIndicator(data['Close']).rsi()
```

```python
data['macd'] = ta.trend.MACD(data['Close']).macd()
```

```python
data['bollinger'] = ta.volatility.BollingerBands(data['Close']).bollinger_hband()
```

```python
data['atr'] = ta.volatility.AverageTrueRange(data['High'],data['Low'],data['Close']).average_true_range()
```

```python
data.info()
```

```python
import numpy as np
```

```python
data['atr'] = data['atr'].replace(0,np.nan).ffill()
```

```python
data['macd'] = data['macd'].bfill()
```

```python
data['rsi'].bfill(inplace=True)
```

```python
data['bollinger_hband'] = ta.volatility.BollingerBands(data['Close']).bollinger_hband()
data['bollinger_lband'] = ta.volatility.BollingerBands(data['Close']).bollinger_lband()
data['bollinger_mavg'] = ta.volatility.BollingerBands(data['Close']).bollinger_mavg()
```

```python
data.drop(columns=['bollinger'], inplace=True)
```

```python
bands = ['hband','lband','mavg']
for band in bands:
    data['bollinger_'+band] = data['bollinger_'+band].bfill()
```

```python
data['atr'] = data['atr'].bfill()
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(16,9))
plt.plot(data['Datetime'],data['Close'], color='blue')
plt.plot(data['Datetime'],data['bollinger_hband'], color = 'red' )
plt.plot(data['Datetime'],data['bollinger_lband'], color = 'green' )
plt.plot(data['Datetime'],data['bollinger_mavg'], color = 'yellow' )
plt.legend()
plt.show()
```
